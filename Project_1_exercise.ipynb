### 2. Manually calculate standard deviation:

```python
# Code:
# 1. Calculate the mean (average) of the dataset.
# 2. Subtract the mean from each data point and square the result (this is called the squared deviation).
# 3. Compute the average of these squared deviations.
# 4. Take the square root of this average to get the standard deviation.
# Sample Data   
import math
data = [10, 12, 23, 23, 16, 23, 21, 16]

#if __name__ == "__main__":
#    main()

def sum_calculate(data):
    return sum(data)

def mean_calculate(data):
    result = sum_calculate(data) / len(data)
    return result

def squared_deviations(mean, data):
        mean_val = mean_manual(data)
        squared_dev = [(x - mean_val) ** 2 for x in data]
        return squared_dev

def variance_calculate(squared_deviations_val, data):
        return sum(squared_deviations_val) / len(data)

def sqrt_calculate(varience_val):
        return math.sqrt(varience_val)

def std_deviation(data):
    # Step 1
    mean_val = mean_calculate(data)
    
    # Step 2
    # squared_deviations_val = [(x - mean_val) ** 2 for x in data]
    squared_deviations_val = squared_deviations(mean_val, data)
    
    # Step 3
    # varience_val = sum(squared_deviations_val) / len(data)
    varience_val = variance_calculate(squared_deviations_val, data)
    
    # Step 4
    # std_deviation_val = math.sqrt(varience_val)
    std_deviation_val = sqrt_calculate(varience_val)
    
    return std_deviation_val

# Recheck Result
# print(sum(data))
# print(mean_val)
# print(squred_deviations_val)

result_value = std_deviation(data)
print("Standard Deviation: ", result_value)
```

```python
def find_duplicate_columns(df):
    duplicates = []
    for i in range(df.shape[1]):
        col1 = df.iloc[:, i]
        for j in range(i + 1, df.shape[1]):
            col2 = df.iloc[:, j]
            if col1.equals(col2):
                duplicates.append((df.columns[i], df.columns[j]))
    return duplicates
```
